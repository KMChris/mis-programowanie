# Zadanie nr 5

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
```

```python
N = 162 # Maksymalna liczba iteracji
px = 1000 # Rozdzielczość wygenerowanego obrazu
Mandelbrot = np.empty([px, px], dtype='?')
mpl.rcParams['figure.dpi'] = 300
```

## Zadanie 1 &ndash; Zbiór Mandelbrota

```python
@np.vectorize
def mandelbrot_cell(z):
    c = 0
    for _ in range(1, N):
        if abs(c := c * c + z) > 2:
            return False
    return True
```

```python
def mandelbrot(real_min=-2, real_max=2, imag_min=-2, imag_max=2):
    global Mandelbrot
    real, imag = np.meshgrid(np.linspace(real_min, real_max, px), np.linspace(imag_min, imag_max, px))
    Mandelbrot = mandelbrot_cell(np.vectorize(complex)(real, imag))
```

```python
mandelbrot()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(Mandelbrot)
ax.set_title("Zbiór Mandelbrota")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
plt.show()
```

## Zadanie 2 &ndash; Zbiory Julii

```python
@np.vectorize
def julia_cell(z, c):
    for _ in range(1, N):
        if abs(z := z * z + c) > 2:
            return False
    return True
```

```python
def julia(c, real_min=-2, real_max=2, imag_min=-2, imag_max=2):
    global Mandelbrot
    real, imag = np.meshgrid(np.linspace(real_min, real_max, px), np.linspace(imag_min, imag_max, px))
    Mandelbrot = julia_cell(np.vectorize(complex)(real, imag), c)
```

```python
julia(complex(-0.1, 0.65))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(Mandelbrot)
ax.set_title(r"$c=-0.1+0.65i$")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
plt.show()
julia(complex(0.285, 0.015))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.matshow(Mandelbrot)
ax.set_title(r"$c=0.285+0.015i$")
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
plt.show()
```

## Powiększenie zbioru Mandelbrota

```python
px=500
Mandelbrot = np.empty([px, px], dtype='?')
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

mandelbrot(-0.75, -0.65, -0.35, -0.25) # Zoom 40x
ax1.matshow(Mandelbrot)
ax1.set_title("Zoom 40 razy", pad=-1)

mandelbrot(-0.7, -0.68, -0.3, -0.28) # Zoom 200x
ax2.matshow(Mandelbrot)
ax2.set_title("Zoom 200 razy", pad=-1)

mandelbrot(-0.7, -0.695, -0.3, -0.295) # Zoom 800x
ax3.matshow(Mandelbrot)
ax3.set_title("Zoom 800 razy", pad=-1)

mandelbrot(-0.7006, -0.7005, -0.2976, -0.2975) # Zoom 40000x
ax4.matshow(Mandelbrot)
ax4.set_title("Zoom 40 tys. razy", pad=-1)

for ax in (ax1, ax2, ax3, ax4):
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)

plt.show()
```
